Visualize the data using Python by plotting the graphs for assignment no. 1 and 2. Consider a suitable data set. Use Scatter plot, Bar plot, Box plot, Pie chart, Line Chart.

```python
import pandas as pd

df = pd.DataFrame({
    "Study_Hours": [1, 2, 3, 4, 5, 6, 7],
    "Attendance": [60, 65, 70, 75, 80, 85, 90],
    "Marks": [50, 55, 60, 65, 72, 78, 85]
})

print(df)
```

Scatter plot

```python
#scatter plot
import matplotlib.pyplot as plt

plt.scatter(df["Study_Hours"], df["Marks"])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()
```

Bar plot

```python
#bar plot
plt.bar(df['Study_Hours'], df['Marks'])
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Bar Plot: Study Hours vs Marks')
plt.show()
```

Box plot

```python
#box plot
plt.boxplot([df['Study_Hours'], df['Attendance'], df['Marks']],
            labels=['Study hours', 'attendance', 'marks'])
plt.title('Box Plot of Study hours, attendance and Marks')
plt.ylabel('Values')
plt.show()
```

Pie chart

```python
#pie chart
import matplotlib.pyplot as plt

average_values = df[['Study_Hours', 'Attendance', 'Marks']].mean()

plt.pie(average_values,
        labels=average_values.index,
        autopct='%1.1f%%')
plt.title('Pie Chart: Average Distribution of Features')
plt.show()
```

Line chart

```python
#line chart
import matplotlib.pyplot as plt

plt.plot(df['Study_Hours'], df['Marks'], marker='o')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Line Graph: Study Hours vs Marks')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(df.index, df['Study_Hours'], label='Study Hours')
plt.plot(df.index, df['Attendance'], label='Attendance')
plt.plot(df.index, df['Marks'], label='Marks')

plt.xlabel('Students')
plt.ylabel('Values')
plt.title('Line Chart: Study, Attendance and Marks Trend')
plt.legend()
plt.show()
```
